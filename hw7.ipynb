<h1 align="center">Checking Questions</h1> 

**Вопрос 1**: Можно ли использовать сверточные сети для классификации текстов? Если нет обоснуйте :D, если да то как? как решить проблему с произвольной длинной входа?

Можно закодировать слова числовыми векторами. Например one hot encoding, но лучше word2vec. Далее вектора слов записывем в матрицу (один под другим в строки). Получили матрицу, характеризующую текст. К ней можно применять сверточные сети. Кажется, эта штука не интуитивно учитывает порядок слов. Сверточные сети для изображений учатся абстрагироваться от расположения объекта на картинке. Но в случае картинок мы для этого предпринимали специльные меры (сдвиги с пэдинком, отражения и т. д.). Если этих мер не предпринимать, как-то сеть учитываться порядок может. Однако это происходит очень сложным образом и вряд начнет проявлять себя без переобучения. Чтобы решить проблему, можно использовать n-граммы. Для решения проблемы произвольности длины входа, можно, например, разбить тексты на куски фиксированной длины, которые влезают в память. Обрабытваем их последовательно от эпохи к эпохе. Тот кусок текста (последний), который получился меньше остальных, можно дополнить специальным символом, означиющим пустоту.

**Вопрос 2**: Чем LSTM лучше/хуже чем обычная RNN?

Использование LSTM спасает от проблемы затухания градиентов. В итоге можно делать сети "глубже смотрящие" в прошлое. Минус LSTM, напрмер, в том, что градиент может взрываться (как описано в следущем вопросе).

**Вопрос 3**:  Выпишите производную $\frac{d c_{n+1}}{d c_{k}}$ для LSTM http://colah.github.io/posts/2015-08-Understanding-LSTMs/, объясните формулу, когда производная затухает, когда взрывается?

$\frac{d c_{n+1}}{d c_{k}} = \frac{d c_{n+1}}{d c_{n}}\frac{d c_{n}}{d c_{k}}=\prod\limits_{i=k+1}^{n+1}diag(f_i)$, где произведение поэлементное. Если это значение больше $1$, градиент взрывается. Иначе нет. Это контролирует forget gate.

**Вопрос 4**: Зачем нужен TBPTT почему BPTT плох?

Так как градиент затухает, влияние его с дальних слоев на ответ крайне мало. Поэтому обновление весов дальних слоев это много лишних, долгих, но почти бесполезных вычислений, которые можно убрать. Это позволит значительно ускорить обучение.

**Вопрос 5**: Как комбинировать рекуррентные и сверточные сети, а главное зачем? Приведите несколько примеров реальных задач.

В нашей задаче (подписывание картинок) выход сверточной сети является информативным вектором признаков для картинки и инициализирует внутреннее состояние. Соответственно в любой другой задаче, в которой требуется каким-либо образом связывать картинки с данными другого типа, можно использовать подобную комбинацию. Например генерирование звуков (синтез речи), видео и т. д.

**Вопрос 6**: Объясните интуицию выбора размера эмбединг слоя? почему это опасное место?

Если размер слишком маленький, не удастся закодировать всю информацию, содержащуюся в тексте. Это приведет к недообучению. Очень большой размер - тоже плохо. Во-первых, это лишний расход памяти. Во-вторых, от увеличения размера эмбединг слоя с некоторого момента вся информация уже будет закодирована, поэтому это увеличение будет бесполезно, а бесполезное увеличение параметров приводит к переобуччению. В этом и состояит интуиция выбора размера. Нужно подобрать такой размер, чтобы он соответствовал количеству информации, которую нужно закодировать.
